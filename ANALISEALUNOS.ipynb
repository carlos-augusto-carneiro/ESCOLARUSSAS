```python
data_01_ordenado = data_01.sort_values(by='NOME COMPLETO')
data_01_ordenado
```

```python
data_02_ordenado = data_02.sort_values(by='NOME COMPLETO')
data_02_ordenado
```

```python
data_03_ordenado = data_03.sort_values(by='NOME COMPLETO')
data_03_ordenado
```

```python
data_04_ordenado = data_04.sort_values(by='NOME COMPLETO')
data_04_ordenado
```

```python
DF_MESCLADO = pd.merge(data_01_ordenado, data_02_ordenado, on='NOME COMPLETO',how='left', suffixes=('_Prova1','_Prova2'))
DF_MESCLADO_02 = pd.merge(DF_MESCLADO, data_03_ordenado, on='NOME COMPLETO', how='left')
DF_MESCLADO_03 = pd.merge(DF_MESCLADO_02, data_04_ordenado, on='NOME COMPLETO', how='left', suffixes=('_Prova3','_Prova4'))

DF_MESCLADO_03
```

```python
nomes_repetidos = DF_MESCLADO_03['NOME COMPLETO'].duplicated().sum()
nomes_repetidos

nomes_repetidos2 = DF_MESCLADO_03['NOME COMPLETO'].value_counts()
print("Nomes duplicados:")
print(nomes_repetidos2)
```

```python
df_NOVO = DF_MESCLADO_03.drop(['Serie_Prova2', 'Serie_Prova3', 'Serie_Prova4'], axis=1)
df_NOVO
```

```python
df_NOVO.rename(columns={'Serie_Prova1': 'Serie'}, inplace=True)
df_NOVO
```

```python
def calcular_media(row):
    notas = [row['Pontuação_Prova1'], row['Pontuação_Prova2'], row['Pontuação_Prova3'], row['Pontuação_Prova4']]
    notas_disponiveis = [pontuacao for pontuacao in notas if pd.notna(pontuacao)]
    return sum(notas_disponiveis) / len(notas_disponiveis) if notas_disponiveis else None

df_NOVO['Media'] = df_NOVO.apply(calcular_media, axis=1)
```
